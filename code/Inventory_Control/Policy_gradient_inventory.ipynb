```python
import numpy as np
np.set_printoptions(formatter={'float': lambda x: "{0:0.6f}".format(x)})
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
```

```python
'''
Parameters: norizon length, demand distribution, start state distribution
'''
np.random.seed(10)
horizon = 20
demand_min = 0
demand_max = 1
start_mean = 1
start_var = 0.5
'''
Costs
'''
c = 0.5
p = 0.6
h = 0.55
```

```python
theta = np.random.uniform(0,1,size=(horizon))
print(theta)
```

```python
'''
Function to simulate a trajectory
Input: current threshold level theta as an array
Output: sequence of states and actions as arrays
'''
def simulate(theta,horizon):
    state = np.zeros(horizon+1)
    action = np.zeros(horizon)
    demand = np.random.uniform(demand_min,demand_max,size=(horizon))
    du_dtheta = np.zeros(horizon+1)
    tau = np.zeros(horizon)
    
    state[0] = np.random.normal(start_mean,start_var)
    for i in range(0,horizon):
        action[i] = max(0,theta[i] - state[i])
        state[i+1] = state[i] + action[i] - demand[i]
    
    du_dtheta = np.sign(action)
    dr = h*(state>0) + (-p)*(state<0)
    
    print('Demand',demand)
    print('Action',action)
    print('State',state)
    print('du_dtheta',du_dtheta)
    print('dr',dr)
    return state, demand, action, du_dtheta, dr
```

```python
state, demand, action, du_dtheta, dr = simulate(theta,horizon)
```

```python
def policy_grad(du_dtheta,dr):
    grad = np.zeros(horizon)
    idx = (np.argwhere(du_dtheta > 0)).flatten()
    print(idx)
    for i in range(len(idx)-1):
        print(dr[idx[i]+1:idx[i+1]+1])
        grad[idx[i]] = np.sum(dr[idx[i]+1:idx[i+1]+1])
    '''
    Gradient of the last order takes into account the inventory at the end of the decision horizon
    '''
    grad[idx[-1]] = c + np.sum(dr[idx[-1]+1:])
    print(grad)
    return grad
```

```python
'''
Single function that simulates the gradient
'''
```

```python
def policy_grad_simulate(theta,horizon):
    '''
    Simulation part
    '''
    state = np.zeros(horizon+1)
    cost = np.zeros(horizon+1)
    action = np.zeros(horizon)
    demand = np.random.uniform(demand_min,demand_max,size=(horizon))
    du_dtheta = np.zeros(horizon+1)
    tau = np.zeros(horizon)
    
    state[0] = np.random.normal(start_mean,start_var)
    for i in range(0,horizon):
        action[i] = max(0,theta[i] - state[i])
        cost[i] = c*action[i] + h*max(0,state[i]) + p*max(0,-state[i])
        state[i+1] = state[i] + action[i] - demand[i]
    
    cost[-1] = h*max(0,state[-1]) + p*max(0,-state[-1])
    
    '''
    Gradient computation
    '''
        
    du_dtheta = np.sign(action)
    dr = h*(state>0) - p*(state<0)
    
    grad = np.zeros(horizon)
    idx = (np.argwhere(du_dtheta > 0)).flatten()
    for i in range(len(idx)-1):
        grad[idx[i]] = np.sum(dr[idx[i]+1:idx[i+1]+1])
        
    grad[idx[-1]] = c + np.sum(dr[idx[-1]+1:])
    return np.sum(cost), grad
    
```

```python
'''
Check
'''
cost, grad = policy_grad_simulate(theta,horizon)
print(cost)
print(grad)
```

## Policy gradient in action

```python
'''
Gradient decent
'''
N = 50000
num_sims = 500
stepsize = 0.005
# Parameters for line search
alpha = 1
beta = 0.5
theta = np.random.uniform(0,1,size=(horizon))
gap = []

for k in range(N):
    
    ell = 0
    gradient = np.zeros(horizon)
    for _ in range(num_sims):
        cost, grad = policy_grad_simulate(theta,horizon)
        gradient += grad
        ell += cost
        
#     print(gradient)
    theta -= stepsize*gradient/num_sims
    
    if k % 100 == 0:
        print('Average cost', ell/num_sims)
#         print('Thresholds',theta)
        gap.append(ell/num_sims)
    
```

```python
f = plt.figure
plt.plot(np.array(gap))
plt.title('Cost during training')
plt.ylabel('Cost')
plt.xlabel('Iteration number/100')
```
